# Data Cleaning 

```python
#Removing extra spaces from the features Using strip()
churn_market.columns = churn_market.columns.str.strip()
```

```python
#Finding Outliers using IQR score

#The IQR is the first quartile subtracted from the third quartile; 
#these quartiles can be clearly seen on a box plot on the data.
# calculate IQR score
#where Q3 is the 75th percentile of the data and Q1 is the 25th percentile of the data.
Q1 = churn_market.quantile(0.25)
Q3 = churn_market.quantile(0.75)
IQR = Q3 - Q1
print(IQR)
# IQR of each column
```

```python
#remove outlier
import warnings
warnings.simplefilter(action="ignore", category=FutureWarning)
#Once the IQR is calculated, we can use it to identify outliers by defining a threshold range as follows;
#Lower threshold = Q1 - 1.5 * IQR
#Upper threshold = Q3 + 1.5 * IQR
#Any data points that fall outside of this range are considered outliers and removed from the dataset

churn_market = churn_market[~((churn_market < (Q1 - 1.5 * IQR)) |(churn_market > (Q3 + 1.5 * IQR))).any(axis=1)]
churn_market.shape
```

```python
#create a list for categorical varables 
cat_vars = ['Churn',
             'Contract',
             'InternetService',
             'Phone',
             'Multiple',
             'OnlineSecurity',
             'OnlineBackup',
             'DeviceProtection',
             'TechSupport',
             'StreamingTV',
             'StreamingMovies',
             'PaperlessBilling',
             'PaymentMethod']
```

```python
#create a list for continouse variable
cont_vars = ['Outage_sec_perweek',
             'Yearly_equip_failure',
             'Tenure',
             'MonthlyCharge',
             'Bandwidth_GB_Year']
```

# Parsing Transaction Database into Apriori Algorithm for Market Bucket Analysis

```python
#Association rules on the new dataframe
apriori(basket_input,min_support=0.2,use_colnames=True).head(10)
```

```python
# Build the Apriori Algorithm for Market Basket Analysis

frequent_itemsets = apriori(basket_input,min_support=0.2,use_colnames=True)
rules = association_rules(frequent_itemsets,metric='lift',min_threshold=0.5)
rules.head(10)


#Here, the “antecedents” and “consequents” columns show items that are frequently purchased together.

#In this Analysis, the first row of the dataset tells us that if a person were to buy month-to-month conract,
#then they are also likely to purchase Fibre Optic Internet Service
```

```python
#To get the most frequent item combinations in the entire dataset, 
#let’s sort the dataset by support, confidence, and lift:

rules.sort_values(["support", "confidence","lift"],axis = 0, ascending = False).head(10)



#The resulting table shows the most popular product combinations that are frequently bought together are:

#PaperlessBilling_Yes and Phone_Yes (One reason for this could be that the Telecom company ran a promotion on these items 
#together or displayed them within the same product category to improve sales)

#OnlineSecurity_No and Phone_Yes
#TechSupport_No and Phone_Yes
#DeviceProtection_No and Phone_Yes
#OnlineBackup_No and Phone_Yes
```
